```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
W = 2560
H = 2560
```

```python
image = np.fromfile("histogram.bin", dtype=np.uint32)
image.shape
```

```python
image = image.reshape((H, W))
image.shape
```

```python
img = (image)
```

```python
plt.imshow(img, cmap='hot')
plt.axis('off')
plt.show()
```

```python
cmaps = ['hsv', 'twilight']
    
# Plot each cmap in a grid
fig, axes = plt.subplots(1, len(cmaps), figsize=(15, 3))
for ax, cmap in zip(axes, cmaps):
    ax.imshow(img, cmap=cmap, origin="lower")
    ax.set_title(cmap)
    ax.axis("off")

plt.tight_layout()
plt.show()
```

```python
saves = ["twilight"]
for save in saves:
    plt.imsave(f"buddhabrot_{save}.png", img , cmap=f'{save}')
```
