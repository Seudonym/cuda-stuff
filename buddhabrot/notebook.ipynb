```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
W = 800
H = W
```

```python
image = np.fromfile("temp/histogram.raw", dtype=np.uint32)
image = image.reshape((H, W))
image.shape
```

```python
image.min(), image.max(), image[image != 0].shape, (image == image.max()).sum()
```

```python
plt.imshow(image)
plt.axis('off')
plt.show()
```

```python
cmaps = ['hsv', 'twilight']
    
# Plot each cmap in a grid
fig, axes = plt.subplots(1, len(cmaps), figsize=(15, 3))
for ax, cmap in zip(axes, cmaps):
    ax.imshow(image, cmap=cmap, origin="lower")
    ax.set_title(cmap)
    ax.axis("off")

plt.tight_layout()
plt.show()
```
